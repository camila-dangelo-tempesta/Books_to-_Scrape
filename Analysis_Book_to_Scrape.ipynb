```python
# Libraries
import pandas as pd
import numpy as npb
```

```python
data_row = pd.read_csv("C:/Users/Utilizador/repos/Python_ds_ao_dev/Projeto_Book_to_Scrape/data_row.csv", sep=",", encoding='ISO-8859-1')
```

```python
data_row.columns
```

```python
data_row.drop(['Unnamed: 0'], axis=1, inplace=True)
```

```python
data_row
```

```python
hip_1 = data_row.filter(items = ['product_price','product_catalog', 'product_star']).groupby(['product_catalog', 'product_star']).median().sort_values(by=['product_price','product_star'], ascending= True).reset_index()
hip_1
```

```python
hip_2 = data_row.filter(items = ['product_price','product_catalog', 'product_star']).groupby(['product_catalog', 'product_star']).min().sort_values(by=['product_price','product_star'], ascending= True).reset_index()
hip_2
```

```python
hip_3 = data_row.filter(items = ['product_price','product_catalog', 'product_star']).groupby(['product_catalog', 'product_star']).max().sort_values(by=['product_price','product_star'], ascending= True).reset_index()
hip_3
```

```python
data_row = data_row.drop_duplicates()
```

```python
data_row.to_csv('data_product_book_cleaned.csv')
```
